```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

Application of convolutions

```python
# create signal
Srate = 256 # Hz
t = np.arange(0, 3, 1/Srate)
```

```python
# create a random signal
x = np.sin(2*np.pi * 2*t)
# create a random noise
noise = 5*np.random.randn(len(t))
# NoisySignal
NoisySignal = x + noise
```

```python
# plotting original signal
plt.figure(figsize=(20, 12))
style.use('dark_background')
plt.plot(t, x, label='Original Signal')
plt.legend(fontsize=30)
plt.show()
```

```python
# plotting original signal
plt.figure(figsize=(20, 12))
style.use('dark_background')
plt.stem(t, NoisySignal, label='Noisy Signal')
plt.legend(fontsize=30)
plt.show()
```

```python
filters = 2*np.ones(50)/10.0
print(filters)
```

```python
filteredSignal = np.convolve(NoisySignal, filters, mode='full')
```

```python
# plotting original signal
plt.figure(figsize=(20, 12))
style.use('dark_background')
plt.plot(filteredSignal, label='Filtered Signal')
plt.legend(fontsize=30)
plt.show()
```

Double filter

```python
filteredSignal = np.convolve(filteredSignal, filters, mode='full')
```

```python
# plotting original signal
plt.figure(figsize=(20, 12))
style.use('dark_background')
plt.plot(filteredSignal, label='Filtered Signal')
plt.legend(fontsize=30)
plt.show()
```

```python
filteredSignal = np.convolve(filteredSignal, filters, mode='full')
```

```python
# plotting original signal
plt.figure(figsize=(20, 12))
style.use('dark_background')
plt.stem(filteredSignal, label='Filtered Signal')
plt.legend(fontsize=30)
plt.show()
```

```python
xn, hn = np.array([1, 2, 3, 1]),  np.array([1, 2, 1, -1])
yn = [i*xn for i in hn]
is_valid = lambda r, c: 0 <= r < len(yn) and 0 <= c < len(yn[0])
def calc(r, c):
    res = 0
    while is_valid(r, c):
      res, r, c = res + yn[r][c], r - 1, c + 1
    return res
res = [calc(i, 0) for i in range(len(yn))] + \
      [calc(len(yn) - 1, i) for i in range(1, len(yn[0]))]
print(res)
```
