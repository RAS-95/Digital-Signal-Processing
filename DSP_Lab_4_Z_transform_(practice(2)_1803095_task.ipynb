```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Z_Transformation():

  z=(complex)(2*(np.exp(0.5j)))
  Z=np.zeros(20, dtype=complex)
  Z_total=(complex)(0)
  for k in range(0,20):
    Z[k]=(0.5)**(k)*z**(-k)
    Z_total+=Z[k]

  roc= 'Entire Z-plane'

  roc = 'ROC : |Z|> {}'.format(np.abs(Z_total))

  return Z_total,roc






Z,roc=Z_Transformation()


#plotting

fig,ax=plt.subplots()
circle = plt.Circle((0,0),radius=np.abs(Z), edgecolor='r')
ax.add_patch(circle)
plt.xlim(-np.max(np.abs(Z))-1,np.max(np.abs(Z))+1)
plt.ylim(-np.max(np.abs(Z))-1,np.max(np.abs(Z))+1)

plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title("ROC")
plt.grid(True,which='both',linestyle='--',linewidth=0.5)
plt.show()

#printing  z value


print(roc)
```
